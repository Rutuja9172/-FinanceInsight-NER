```python
text = "RBI approved a loan of 5 crore rupees to Infosys in 2024"
doc = nlp(text)

for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
def extract_entities(text):
    doc = nlp(text)
    entities = {
        "ORGANIZATION": [],
        "MONEY": [],
        "DATE": []
    }

    for ent in doc.ents:
        if ent.label_ == "ORG":
            entities["ORGANIZATION"].append(ent.text)
        elif ent.label_ in ["MONEY", "DATE"]:
            entities[ent.label_].append(ent.text)

    return entities
```

```python
extract_entities("Infosys received 10 crore rupees funding in 2025")
```

```python
import pandas as pd

data = {
    "Text": [
        "RBI approved a loan of 5 crore rupees to Infosys in 2024",
        "HDFC Bank sanctioned 10 lakh rupees in 2023",
        "TCS received funding of 8 crore rupees in 2025"
    ]
}

df = pd.DataFrame(data)
df
```

```python
results = []

for text in df["Text"]:
    entities = extract_entities(text)
    results.append({
        "Text": text,
        "Organizations": entities["ORGANIZATION"],
        "Money": entities["MONEY"],
        "Date": entities["DATE"]
    })

result_df = pd.DataFrame(results)
result_df
```
